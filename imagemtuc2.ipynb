```python
tucano = Image.open('/content/Captura de tela 2022-10-08 184230.png')
plt.imshow(tucano)
```

# Exibir o canal Red em tons de vermelho

```python
tucanoArray = np.array(tucano)

canal_red = tucanoArray[:,:,0]
canal_green = tucanoArray[:,:,1]
canal_blue = tucanoArray[:,:,2]

# Zerando os canais verde e azul
canal_green *= 0
canal_blue *= 0

plt.imshow(tucanoArray)
```
