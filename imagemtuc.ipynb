```python
from PIL import Image, ImageOps

img = Image.open("/content/Captura de tela 2022-10-08 184230.png")

width, height = img.size#pega a largura e a altura da imagem

img = img.resize((width, height))#função para redimensionar a imagem, não foi necessária nesse caso

img_RBGAbands = Image.Image.split(img)#Separa as imagens nas bandas RGBA, mas deixa sem as cores, apenas separa

display(img)

img_Rband = img_RBGAbands[0]
img_Gband = img_RBGAbands[1]
img_Bband = img_RBGAbands[2]

img_Rband = ImageOps.colorize(img_Rband,black="black", white="red")
img_Gband = ImageOps.colorize(img_Gband,black="black", white="green")
img_Bband = ImageOps.colorize(img_Bband,black="black", white="blue")

display(img_Rband)
display(img_Gband)
display(img_Bband)
```
